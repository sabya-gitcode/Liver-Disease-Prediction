# Encoding

## Gender is a categorical feature with values Male and Female

## To deal with this, we convert into numeric form using Label Encoding Or One-Hot Encoding

## Our goal is to assign Male with Encoded value '1' and Female with Encoded value '2'.

```python
encoder = LabelEncoder()
```

```python
df['Gender'] = encoder.fit_transform(df['Gender'])
df.head(2)
```

```python
# We observe the output of df.head() to confirm the encoding process.
```

```python
# We now move onto scaling the feature.
```

# Feature Scaling

```python
# Our goal is to make sure that our algorithm does not treat larger numbers as more important.
# To do so, we scale all features/columns, except for the target column.
```

```python
features = ["Age", "Total_Bilirubin",	"Direct_Bilirubin",	"Alkaline_Phosphotase",	"Alamine_Aminotransferase",	"Aspartate_Aminotransferase",	"Total_Protiens",	"Albumin",	"Albumin_and_Globulin_Ratio"]

scaler = StandardScaler()

df[features] = scaler.fit_transform(df[features]) # this is to scale all the features we defined in variable features through standard scaling.
```

```python
df.head(3)
```

```python
# Next, we balance the data, so the classes are balanced so one class is
# not given more importance over the others.
```

# Data Balancing

```python
from imblearn.over_sampling import SMOTE # SMOTE replicates the minority classes
# until the minority and majority classes are equally balanced.

X = df.drop('Dataset' ,axis = 1)
y = df['Dataset']

smote = SMOTE()

# X_resampled constitutes the input features
# y_resampled constitutes the target features

X_resampled, y_resampled = smote.fit_resample(X,y)

y_resampled.value_counts()

```

# Splitting The Data Into Training And Testing Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size = 0.2, random_state = 42)
```

# Measuring Predictions Through Logistic Regression, Random Forest, SVC

# Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

model = LogisticRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("\n\n", classification_report(y_test, y_pred))
print("\n\n", confusion_matrix(y_test, y_pred))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report

model_r = RandomForestClassifier()

model_r = model_r.fit(X_train, y_train)

y_pred = model_r.predict(X_test)

print("\n\n", classification_report(y_test, y_pred))
print("\n\n", confusion_matrix(y_test, y_pred))
```

# SVC

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report

model = SVC()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("\n\n", classification_report(y_test, y_pred))
print("\n\n", confusion_matrix(y_test, y_pred))
```
